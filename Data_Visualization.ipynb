```python
import pandas as pd
import seaborn as sns
```

```python
# LinePlot
# Initialize data of lists
data = {'Name': ['Mohe', 'Karnal', 'Yrik', 'jack'],
        'Age': [30, 21, 29, 28]}
df = pd.DataFrame(data)
# Create a new column 'Weight' with sample data
df['Weight'] = [70, 65, 72, 68]
# Plotting lineplot
sns.lineplot(x='Age', y='Weight', data=df)
```

```python
# Scatter Plot
# Initialize data of lists
data = {'Name': ['Mohe', 'Karnal', 'Yrik', 'jack'],
        'Age': [30, 21, 29, 28]}
df = pd.DataFrame(data)

# Create a new column 'Weight' with sample data
df['Weight'] = [80, 65, 72, 68]

# Create a scatter plot
sns.scatterplot(x='Age', y='Weight', data=df)
```

```python
# Box Plot
data = {'Name':[ 'Mohe' , 'Karnal' , 'Yrik' , 'jack' ],
        'Age':[ 30 , 21 , 29 , 28 ]}
df = pd.DataFrame( data )
sns.boxplot( data['Age'] )
```

```python
# Box Plot with More Features

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/nba.csv')
sns.boxplot(x='Age', y='Weight',hue='Position',data=df)
```

```python
# Violin Plot
data = {'Name':[ 'Mohe' , 'Karnal' , 'Yrik' , 'jack' ],
        'Age':[ 30 , 21 , 29 , 28 ]}
df = pd.DataFrame( data )
sns.violinplot(data['Age'])
```

```python
# Swarm Plot
sns.set(style = 'whitegrid')

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/nba.csv')
sns.swarmplot(x='Age',hue='Position',data=df)
```

```python
# BarPlot
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/nba.csv')
sns.barplot(x='Position',y='Weight', hue='Age',data=df)
```

```python
# Point Plot
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/nba.csv')
sns.pointplot(x='Age',y='Weight', hue='Position',data=df)
```

```python
# Count Plot
sns.set(style = 'whitegrid')
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/nba.csv')
sns.countplot(x='Age',data=df)
```

```python
# importing the required libraries
from sklearn import datasets
import pandas as pd
import seaborn as sns

# Setting up the Data Frame
iris = datasets.load_iris()

iris_df = pd.DataFrame(iris.data, columns=['Sepal_Length',
                      'Sepal_Width', 'Patal_Length', 'Petal_Width'])

iris_df['Target'] = iris.target

iris_df['Target'].replace([0], 'Iris_Setosa', inplace=True)
iris_df['Target'].replace([1], 'Iris_Vercicolor', inplace=True)
iris_df['Target'].replace([2], 'Iris_Virginica', inplace=True)

# Plotting the KDE Plot
# Changed 'Label' to 'label'
sns.kdeplot(iris_df.loc[(iris_df['Target'] =='Iris_Virginica'), 'Sepal_Length'], color = 'b', fill = True, label ='Iris_Virginica')
```

```python
# KDE Plot for Age and Number feature
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/nba.csv').head()
sns.kdeplot(x='Age', y='Number',data = df)
```

```python
# Bivariate & Univariate Data
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/nba.csv').head()
sns.boxplot(x='Age', y='Height',data = df)
```
